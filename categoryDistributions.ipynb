```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (your data loading and other code) ...

# --- Function to add percentage labels ---
def add_count_percentage_labels(ax, total, h = 15, s = 10):
    for p in ax.patches:
        count = p.get_height().astype(int)
        percentage = '{:.1f}%'.format(100 * count / total)
        label = f'n={count} ({percentage})'
        x = p.get_x() + p.get_width() / 2
        y = p.get_y() + p.get_height() + h
        ax.annotate(label, (x, y), size=s, ha='center')

# --- Gender Distribution ---
ax = sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 15)  # Add labels for Gender
plt.show()

# --- Age Distribution ---
ax = sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution')
plt.ylabel('Count')
plt.show()


# --- Working Professional vs. Student ---
ax = sns.countplot(x='Working Professional or Student', data=df)
plt.title('Working Professional vs. Student')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
add_count_percentage_labels(ax, len(df), 15)  # Add labels for Work/Student
plt.show()

# --- Profession Distribution ---
ax = sns.countplot(y='Profession', data=df, order=df['Profession'].value_counts().index)
plt.title('Profession Distribution')
plt.yticks(fontsize=8)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / len(df))
    x = p.get_width()
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), size=8, ha='left', va='center')
plt.show()

# --- Depression Distribution ---
ax = sns.countplot(x='Depression', data=df)
plt.title('Depression Distribution')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 15)
plt.show()

# --- Academic Pressure Distribtuion ---
ax = sns.countplot(x='Academic Pressure', data=df)
plt.title('Academic Pressure Distribution (n=502)')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 2, 8)
plt.show()

# --- Work Pressure Distribtuion ---
ax = sns.countplot(x='Work Pressure', data=df)
plt.title('Work Pressure Distribution (n=2054)')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 5, 8)
plt.show()

# --- Study Satisfaction Distribtuion ---
ax = sns.countplot(x='Study Satisfaction', data=df)
plt.title('Study Satisfaction Distribution (n=502)')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 1, 8)
plt.show()

# --- Job Satisfaction Distribtuion ---
ax = sns.countplot(x='Job Satisfaction', data=df)
plt.title('Job Satisfaction Distribution (n=2054)')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 5, 8)
plt.show()



```

```python
import numpy as np

sleep_hours_categories = ['Less than 5 hours', '5-6 hours', '7-8 hours', 'More than 8 hours']
category_mapping = {category: i for i, category in enumerate(sleep_hours_categories)}
df['Numeric Sleep Duration'] = df['Sleep Duration'].map(category_mapping)
ax = sns.countplot(x='Numeric Sleep Duration', data=df)
plt.title('Distribution of Sleep Hours')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 8, 8)
ax.set_xticks(np.arange(len(sleep_hours_categories)))
ax.set_xticklabels(sleep_hours_categories, size=8)
plt.show()
```

```python
diet_hours_categories = ['Unhealthy', 'Moderate', 'Healthy']
category_mapping = {category: i for i, category in enumerate(diet_hours_categories)}
df['Numeric Diet'] = df['Dietary Habits'].map(category_mapping)
ax = sns.countplot(x='Numeric Diet', data=df)
plt.title('Distribution of Dietary Habits')
plt.ylabel('Count')
add_count_percentage_labels(ax, len(df), 8, 8)
ax.set_xticks(np.arange(len(diet_hours_categories)))
ax.set_xticklabels(diet_hours_categories, size=8)
plt.show()
```
